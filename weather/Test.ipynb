Because we only want days when there is no weather or obstructions to visibility, we are only interested in data with these weather values:
<br>
- Clear 
- Mainly clear 
- Mostly cloudy
- Cloudy data 

```python
filename = "hourly_2014_2019/2019/en_climate_hourly_BC_1108395_01-2019_P1H.csv"
df = pd.read_csv(filename, index_col=None, header=0, parse_dates = ["Date/Time"])

#Get date time and weather status
filterDf = df[["Date/Time","Year", "Month", "Day", "Time", "Weather"]]
filterDf
```

```python
#Add a column for Date only
filterDf["Date"] = filterDf['Date/Time'].apply(lambda x:x.date().strftime('%Y-%m-%d'))
filterDf
```

```python
#Drop na entries
#Ensures there exist at least 1 weather record for each day in the current month
dfDropna = filterDf
assert(len(dfDropna.Date.unique()) == 31)
```

```python
test = dfDropna[dfDropna["Date"] == "2019-01-01"]
test
```

```python
#determine the weather of the day by finding the most weather occurence
#https://stackoverflow.com/questions/48590268/pandas-get-the-most-frequent-values-of-a-column
test["Weather"].value_counts().idxmax()
```
